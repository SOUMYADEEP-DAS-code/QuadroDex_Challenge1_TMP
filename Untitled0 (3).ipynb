## Customer segmentation

### Subtask:
Create customer segments based on demographics (age, gender), income, and potentially purchasing habits.


**Reasoning**:

```python

segments = df.groupby(['Gender','AgeGroup','Income']).agg(
    TotalCustomers = ('Customer_ID','nunique'),
    AvgPurchase = ('Total_Purchases','mean'),
    TotalRevenue = ('Total_Purchases','sum')
).reset_index()

print(segments.sort_values('TotalRevenue', ascending=False))
```

## Analyze spending habits

### Subtask:

```python

customer_total_spend = df.groupby('Customer_ID')['Total_Amount'].sum().reset_index(name='TotalCustomerSpend')
customer_avg_spend = df.groupby('Customer_ID')['Total_Amount'].mean().reset_index(name='AverageCustomerSpend')


segment_spending = df.groupby(['Gender', 'AgeGroup', 'Income']).agg(
    TotalRevenue = ('Total_Amount', 'sum'),
    AveragePurchaseAmount = ('Total_Amount', 'mean'),
    TotalCustomers = ('Customer_ID', 'nunique') # Also useful to see segment size
).reset_index()

print("Segment Spending Analysis (Sorted by Total Revenue):")
display(segment_spending.sort_values('TotalRevenue', ascending=False))

print("\nSegment Spending Analysis (Sorted by Average Purchase Amount):")
display(segment_spending.sort_values('AveragePurchaseAmount', ascending=False))
```

## Analyze customer loyalty

### Subtask:
Examine purchase frequency and total purchase counts to understand customer loyalty.


**Reasoning**:
The previous code cells failed to calculate purchase frequency and total spend per customer correctly due to issues with accessing or processing the data. I need to re-calculate these metrics using the cleaned `df` DataFrame and then merge them.



```python
purchase_frequency = df.groupby('Customer_ID')['Transaction_ID'].nunique().reset_index(name='PurchaseCount')

customer_total_spend = df.groupby('Customer_ID')['Total_Amount'].sum().reset_index(name='TotalCustomerSpend')

customer_loyalty = pd.merge(purchase_frequency, customer_total_spend, on='Customer_ID')

display(customer_loyalty.head())
```

**Reasoning**:
Now that the `customer_loyalty` DataFrame is created, I need to calculate and display descriptive statistics for 'PurchaseCount' and 'TotalCustomerSpend' to understand their distributions, identify the top 10 customers by purchase count and total spend, and analyze the relationship between the two metrics by calculating their correlation.



```python
print("Descriptive statistics for PurchaseCount:")
display(customer_loyalty['PurchaseCount'].describe())

print("\nDescriptive statistics for TotalCustomerSpend:")
display(customer_loyalty['TotalCustomerSpend'].describe())

print("\nTop 10 customers by Purchase Count:")
display(customer_loyalty.sort_values(by='PurchaseCount', ascending=False).head(10))

print("\nTop 10 customers by Total Spend:")
display(customer_loyalty.sort_values(by='TotalCustomerSpend', ascending=False).head(10))

correlation = customer_loyalty['PurchaseCount'].corr(customer_loyalty['TotalCustomerSpend'])
print(f"\nCorrelation between Purchase Count and Total Customer Spend: {correlation}")
```
